```python
import nltk
import numpy
import matplotlib

nltk.download('punkt')

from nltk.corpus import PlaintextCorpusReader
corpus_root = "assignment1"

text = PlaintextCorpusReader(corpus_root, '.*', encoding = "utf8")
story = text.words('PartyGirls.txt')
news = text.words('PretendianInvestigation.txt')

# LENGTH of each text
print("The length of PartyGirls.txt is: " + str(len(story)))
print("The length of PretendianInvestigation.txt is: " + str(len(news)) + "\n")

# LEXICAL DIVERSITY of each text
print(set(story))
print("\n")
print(sorted(set(w.lower() for w in story)))
print("\n")
sorted(set(w.lower() for w in story if w.isalpha()))
print("\n")
print(set(news))
print("\n")
print(sorted(set(a.lower() for a in story)))
print("\n")
sorted(set(a.lower() for a in news if a.isalpha()))

def lexical_diversity(text):
    return len(set(text)) / len(text)

print("The lexical diversity of PartyGirls.txt is: " + str(lexical_diversity(story)))
print("The lexical diversity of PretendianInvestigation.txt is: " + str(lexical_diversity(news)))

# LONGEST SENTENCE of each text

storySen = text.sents('PartyGirls.txt')
longest1 = max(len(s) for s in storySen)
fullsen1 = [s for s in storySen if len(s) == longest1]
print(fullsen1)
print('\nThe length of the longest sentence is: ' + str(longest1))
```

```python
# TOP COLLOCATIONS of each text

from nltk.collocations import *
bigrams = nltk.collocations.BigramAssocMeasures()
trigrams = nltk.collocations.TrigramAssocMeasures()
quadgrams = nltk.collocations.QuadgramAssocMeasures()
biFinder = BigramCollocationFinder.from_words(w.lower() for w in story if w.isalpha())
triFinder = TrigramCollocationFinder.from_words(w.lower() for w in story if w.isalpha())
quadFinder = QuadgramCollocationFinder.from_words(w.lower() for w in story if w.isalpha())

print(biFinder.nbest(bigrams.pmi, 10))
print("\n")
print(triFinder.nbest(trigrams.pmi, 10))
print("\n")
print(quadFinder.nbest(quadgrams.pmi, 10))
print("\n\n")

# TOP 10 WORDS STARTING WITH EACH VOWEL of each text

vowWords = []
fdist = FreqDist(vowWords)
for j in (w.lower() for w in story if w.isalpha()):
    if j.startswith('a'):
            vowWords.append(j)
fdist = FreqDist(vowWords)
print(fdist.most_common(10))
vowWords.clear()
for j in (w.lower() for w in story if w.isalpha()):
    if j.startswith('e'):
            vowWords.append(j)
fdist = FreqDist(vowWords)
print(fdist.most_common(10))
vowWords.clear()
for j in (w.lower() for w in story if w.isalpha()):
    if j.startswith('i'):
            vowWords.append(j)
fdist = FreqDist(vowWords)
print(fdist.most_common(10))
vowWords.clear()
for j in (w.lower() for w in story if w.isalpha()):
    if j.startswith('o'):
            vowWords.append(j)
fdist = FreqDist(vowWords)
print(fdist.most_common(10))
vowWords.clear()
for j in (w.lower() for w in story if w.isalpha()):
    if j.startswith('u'):
            vowWords.append(j)
fdist = FreqDist(vowWords)
print(fdist.most_common(10))
vowWords.clear()
              
fdist = FreqDist(vowWords)
for j in (w.lower() for w in news if w.isalpha()):
    if j.startswith('a'):
            vowWords.append(j)
fdist = FreqDist(vowWords)
print(fdist.most_common(10))
vowWords.clear()
for j in (w.lower() for w in news if w.isalpha()):
    if j.startswith('e'):
            vowWords.append(j)
fdist = FreqDist(vowWords)
print(fdist.most_common(10))
vowWords.clear()
for j in (w.lower() for w in news if w.isalpha()):
    if j.startswith('i'):
            vowWords.append(j)
fdist = FreqDist(vowWords)
print(fdist.most_common(10))
vowWords.clear()
for j in (w.lower() for w in news if w.isalpha()):
    if j.startswith('o'):
            vowWords.append(j)
fdist = FreqDist(vowWords)
print(fdist.most_common(10))
vowWords.clear()
for j in (w.lower() for w in news if w.isalpha()):
    if j.startswith('u'):
            vowWords.append(j)
fdist = FreqDist(vowWords)
print(fdist.most_common(10))
vowWords.clear()
```
